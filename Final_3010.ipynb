```python
import pandas as pd

# Reading the CSV files
df1 = pd.read_csv('/content/yelp_review_arizona.csv')
df2 = pd.read_csv('/content/yelp_business.csv')

# Merging the CSV files based on one column 'business_id'
df3 = pd.merge(df1, df2, on = 'business_id')
df3.set_index('business_id', inplace = True)

# Write it to a new CSV file
df3.to_csv('CSV3.csv')
df3
```

```python
dataFrame = df3.loc[:, ['name','text', 'attributes','categories','city','state','stars_y']]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string


def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(tokens)


dataFrame['cleaned_text'] = dataFrame['text'].apply(preprocess_text)
```

```python
#Finding key words
def is_third_place(df_x, query_words):
    # To find out if it is false
    df_x['is_third_place'] = False

    for query_word in query_words:
#true or false columns
        df_x.loc[df_x['text'].str.contains(query_word, case=False), 'is_third_place'] = True

    return df_x

# Example keywords
third_place_keywords = ["community","social", "gathering", "local", "neighborhood", "hang out"]
df_data = is_third_place(dataFrame, third_place_keywords)
```

```python
from tabulate import tabulate
import ast

# Filter the table so it finds all the is_third_place true
third_place_df = df_data[df_data['is_third_place'] == True]

# Create a list of lists to store the data for tabulate
table_data = []

# create a list with name, business_id, and the attributes
for index, row in third_place_df.iterrows():
    name = row['name']
    business_id = index
    attributes_dict = ast.literal_eval(row['attributes'])
    # Extract attributes that are True
    attributes = ", ".join([attr for attr, value in attributes_dict.items() if value == 'True'])
    table_data.append([name, business_id, attributes])

print(tabulate(table_data, headers=['Name', 'Business ID', 'Attributes'], tablefmt='grid'))


```

```python
from tabulate import tabulate
import ast

third_place_df = df_data[df_data['is_third_place']]

table_data = []

prev_name = None
prev_business_id = None
prev_attributes = None

# Create a list with name, business_id, and the attributes
for index, row in third_place_df.iterrows():
    name = row['name']
    business_id = index
    attributes_dict = ast.literal_eval(row['attributes'])
    # Extract attributes that are True
    attributes = ", ".join([attr for attr, value in attributes_dict.items() if value == 'True'])

    if name != prev_name or business_id != prev_business_id or attributes != prev_attributes:
        table_data.append([name, business_id, attributes])

    prev_name = name
    prev_business_id = business_id
    prev_attributes = attributes

print(tabulate(table_data, headers=['Name', 'Business ID', 'Attributes'], tablefmt='grid'))




```

```python
import matplotlib.pyplot as plt

attribute_counts = {}

for row in table_data:
    attributes = row[2].split(", ")
    for attribute in attributes:
        if attribute in attribute_counts:
            attribute_counts[attribute] += 1
        else:
            attribute_counts[attribute] = 1

# Convert the dictionary to lists for plotting
attributes = list(attribute_counts.keys())
counts = list(attribute_counts.values())

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.bar(attributes, counts)
plt.xlabel('Attributes')
plt.ylabel('Count')
plt.title('Histogram of Attributes')
plt.xticks(rotation=45, ha='right')
plt.show()

# Print the count
for attribute, count in attribute_counts.items():
    print(f"{attribute}: {count}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Assume df contains the preprocessed data with 'cleaned_text' column and 'is_social_community' labels

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df_data['cleaned_text'], df_data['is_third_place'], test_size=0.2, random_state=42)

# Feature extraction using TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # You can adjust max_features as needed
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

#logistic regression classifier
classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train_tfidf, y_train)

# Evaluate the model
y_pred = classifier.predict(X_test_tfidf)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))

new_reviews = ["This local community center is fantastic!",
               "Great place to hang out with friends.",
               "The food here is delicious but not really community-oriented."]
new_reviews_tfidf = tfidf_vectorizer.transform(new_reviews)
predictions = classifier.predict(new_reviews_tfidf)
print("Predictions for new reviews:", predictions)

```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

#confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat, display_labels=classifier.classes_)
disp.plot()
plt.title('Confusion Matrix for LogisticRegression Model')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
# Train a random forest classifier
classifier = RandomForestClassifier(random_state = 42)
classifier.fit(X_train_tfidf, y_train)

# Evaluate the model
y_pred = classifier.predict(X_test_tfidf)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))

new_reviews = ["This local community center is fantastic!",
               "Great place to hang out with friends.",
               "The food here is delicious but not really community-oriented."]
new_reviews_tfidf = tfidf_vectorizer.transform(new_reviews)
predictions = classifier.predict(new_reviews_tfidf)
print("Predictions for new reviews:", predictions)

#confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat, display_labels=classifier.classes_)
disp.plot()
plt.title('Confusion Matrix for Random Forest Model')
plt.show()
```

```python
import folium

df_data['city'] = df_data['city'].str.lower()

third_place_data = df_data[df_data['is_third_place']]
cities = third_place_data['city'].unique()

city_groups = third_place_data.groupby('city')

city_counts = city_groups.size().reset_index(name='Third_Place_Count')

import matplotlib.pyplot as plt

city_counts = third_place_data['city'].value_counts()
top_cities = city_counts.nlargest(5)

#bar graph
plt.figure(figsize=(10, 6))
top_cities.plot(kind='bar', color='grey')

plt.title('Top Cities with Most Third Places')
plt.xlabel('City')
plt.ylabel('Number of Third Places')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# cities and their top 5 third places

top_five_third_places = {}

for city, group in city_groups:
    # drop duplicates based on 'name' column to consider each business name once
    unique_names = group.drop_duplicates(subset=['name'])
    # Select the top 5 third places based on 'stars_y' column
    top_five_places = unique_names.nlargest(5, 'stars_y')[['name', 'stars_y']]
    top_five_third_places[city] = top_five_places



#top third places for Cities
for city, places in top_five_third_places.items():
    print(f"City: {city}")
    print(tabulate(places, headers='keys', tablefmt='grid'))
    print()
```
